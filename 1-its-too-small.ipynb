# Manually overriding the content and code font size

There's an easily overlooked menu item:



Changing the font size one point at a time is annoying; open the Advanced Settings Editor.



Once you start changing the font size with the menu, Jupyter shows you how to modify your user preferences. Now you can set these to any number.

If you're planning on commenting them out when done, beware! Jupyter likes to _overwrite_ this file.

# Preferences in the notebook

Only presenters need to make their fonts big, so that's why we've been changing the preferences manually.

But what if you'll be running this on a server and don't want to manually reset the preferences on each new machine?

Jupyter is magical: it can change its own CSS (HTML stylesheets).

````python
import IPython.core.display

IPython.core.display.HTML("""
<style>
  /* resize the input editor and output area */
  div.jp-Editor div { font-size: 25px; }
  div.jp-OutputArea-output pre { font-size: 25px; }

  /* resize the markdown output */
  div.jp-MarkdownOutput { font-size: 25px; }
</style>
""")
````

Now you can bind your big-font preferences to the notebook, but...

   * evaluating that cell changes font sizes in all notebooks
   * the fonts will be big as long as the invisible `IPython.core.display.HTML` is in the output of the cell (try "Edit > Clear All Outputs")
   * if your students evaluate it, they might not like what you've done to their Jupyter environment
   * commenting out the `IPython.core.display.HTML` element and re-evaluating the cell resets the style

Maybe you want to put your Jupyter-manipulation code in a separate notebook named `_do_not_run.ipynb`.
