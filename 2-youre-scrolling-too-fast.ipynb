# Solution #2: CSS hacking

The following will make all cells disappear except the currently selected one.

Before evaluating that cell, be aware that

  * `j` goes to the next slide
  * `k` goes to the previous slide
  * you always want to `ctrl-enter`, not `shift-enter`!

```python
import IPython.core.display

IPython.core.display.HTML("""
<style>
  .jp-NotebookPanel-notebook > .jp-Cell:not(.jp-mod-selected) { display: none; }
</style>
""")

# As before, you can disable this by commenting out and re-evaluating this cell.
```
