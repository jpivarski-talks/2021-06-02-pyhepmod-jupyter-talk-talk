```python
import pandas as pd
```

```python
df = pd.read_csv("trends-jupyter-ipython.csv", header=1, parse_dates=[0], names=["date", '"Mathematica"', '"SageMath"', '"Google Colab"', '"iPython"', '"Jupyter"'], na_values="<1").set_index("date")
df
```

```python
ax = df[['"Mathematica"', '"Jupyter"', '"iPython"', '"Google Colab"', '"SageMath"']].fillna(0.001).rolling(3).sum().plot(logy=True, figsize=(8, 4))
ax.set_ylim(0.5, 400)
ax.set_ylabel("Relative Google search volume")
ax.set_xlabel("")
ax.grid(axis="y", which="both", color="lightgray")
ax.legend(loc=(1.02, 0.64), framealpha=1)
ax.get_figure().savefig("img/trends-jupyter-ipython.pdf")
```
